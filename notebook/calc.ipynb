```julia
model = TFIML()
H = model(12)
(, v) = eigen(H)

E = [2:end] - [1:end-1]

display(scatter([1:length()20], ylabel="Eigenvalues", xlabel="Index", title="Eigenvalues of the Hamiltonian"))
histogram(E, bins=30, xlabel="Energy Gap", ylabel="Frequency", title="Histogram of Energy Gaps")
```

```julia
model = XYZ(Jx=1.0, Jy=1.0, Jz=0.8, hx=0.0, hy=0.0, hz=0.1)
H = model(8)
(, v) = eigen(H)

E = [2:end] - [1:end-1]
range = minimum(E):0.01:maximum(E)
poisson_dist = [poisson(E) for E in range]
wigner_dist = [wigner_surmise(E) for E in range]

p_spectrum = plot(xlabel="", ylabel="Energy Level", title="Energy Spectrum")
scatter!(p_spectrum, , label="Poisson Distribution", lw=2)
display(p_spectrum)

p_level = plot(xlabel="Energy Gap", ylabel="Frequency", title="Level Spacing Distribution")
plot!(p_level, range, poisson_dist, label="Poisson Distribution", lw=2)
plot!(p_level, range, wigner_dist, label="Wigner Surmise", lw=2)
histogram!(p_level, E, bins=30, normalize=true, label="Numerical Data", alpha=0.5)
```
