```julia
# TODO : Diagonalize superblock hamiltonian
using DiagonalizeHamiltonian
using SparseArrays, LinearAlgebra

model = TFIM(boundary=PBC())
J, h = model.J, model.h
Hfull = model(4)

abstract type AbstractDMRG end
@kwdef struct PauriMatrix <: AbstractDMRG
    I::SparseMatrixCSC{Float64,Int64} = sparse(LinearAlgebra.I, 2, 2)
    σx::SparseMatrixCSC{ComplexF64,Int64} = sparse([0.0 1.0; 1.0 0.0])
    σy::SparseMatrixCSC{ComplexF64,Int64} = sparse([0.0 -im; im 0.0])
    σz::SparseMatrixCSC{ComplexF64,Int64} = sparse([1.0 0.0; 0.0 -1.0])
end
struct System <: AbstractDMRG
    site::Int

    H_s::SparseMatrixCSC{ComplexF64,Int64}
end
struct Block <: AbstractDMRG
    site::Int
    I_b::SparseMatrixCSC{Float64,Int64}
    σx_b::SparseMatrixCSC{ComplexF64,Int64}
    σy_b::SparseMatrixCSC{ComplexF64,Int64}
    σz_b::SparseMatrixCSC{ComplexF64,Int64}
    H_b::SparseMatrixCSC{ComplexF64,Int64}
    m_trunc::Int
    function Block(system_size::Int, σx_b, σy_b, σz_b, H_b, m::Int)
        @assert size(σx_b) == (m, m)
        @assert size(σy_b) == (m, m)
        @assert size(σz_b) == (m, m)
        @assert size(H_b) == (m, m)
        I_b = sparse(LinearAlgebra.I, m, m)
        new(system_size, I_b, σx_b, σy_b, σz_b, H_b, m)
    end
end
struct SuperBlock <: AbstractDMRG
    left::Block
    system::Vector{System}
    right::Block
end
struct FDMRGState <: AbstractDMRG
    total_sites::Int
    left::Vector{Block}
    right::Vector{Block}
end
Block(1, σx, σy, σz, sparse(zeros(2, 2)), 2)
```
